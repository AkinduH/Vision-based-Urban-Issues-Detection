# Damaged Road Signs

```python
import os
from collections import Counter

folder_path = "C:\\Users\\ravin\\OneDrive\\Documents\\DamagedRoadSigns2\\train\\labels"
label_counts = Counter()

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if parts:  # skip empty lines
                    label = parts[0]
                    label_counts[label] += 1

# Print results
for label, count in sorted(label_counts.items(), key=lambda x: int(x[0])):
    print(f"Label {label}: {count} instances")
```

```python
import os

folder_path = "C:\\Users\\ravin\\OneDrive\\Documents\\DamagedRoadSigns2\\train\\labels"
files_with_label1 = []

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        modified_lines = []
        has_label1 = False
        
        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if parts:
                    # If label is 0, change it to 3
                    if parts[0] == "0":
                        parts[0] = "3"
                    # Check if label 1 exists
                    if parts[0] == "1":
                        has_label1 = True
                    modified_lines.append(" ".join(parts))
        
        # Save changes back to file
        with open(file_path, "w") as f:
            f.write("\n".join(modified_lines) + "\n")
        
        # Add to the list if label 1 was found
        if has_label1:
            files_with_label1.append(filename)

print(files_with_label1)
```

```python
import os

txt_folder = "C:\\Users\\ravin\\OneDrive\\Documents\\DamagedRoadSigns2\\train\\labels"
jpg_folder = "C:\\Users\\ravin\\OneDrive\\Documents\\DamagedRoadSigns2\\train\\images"

for txt_file in files_with_label1:
    txt_path = os.path.join(txt_folder, txt_file)
    jpg_file = os.path.splitext(txt_file)[0] + ".jpg"
    jpg_path = os.path.join(jpg_folder, jpg_file)
    
    # delete txt
    if os.path.exists(txt_path):
        os.remove(txt_path)
        print(f"Deleted TXT: {txt_file}")
    
    # delete jpg
    if os.path.exists(jpg_path):
        os.remove(jpg_path)
        print(f"Deleted JPG: {jpg_file}")
```

# Illegal Parking

```python
import os
from collections import Counter

folder_path = "C:\\Users\\ravin\\OneDrive\\Documents\\IllegalParking\\train\\labels"
label_counts = Counter()

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if parts:  # skip empty lines
                    label = parts[0]
                    label_counts[label] += 1

# Print results
for label, count in sorted(label_counts.items(), key=lambda x: int(x[0])):
    print(f"Label {label}: {count} instances")
```

```python
import os

folder_path = "C:\\Users\\ravin\\OneDrive\\Documents\\IllegalParking\\train\\labels"
files_with_label1 = []

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "r") as f:
            lines = [line.strip().split() for line in f if line.strip()]

        if not lines:
            continue

        if len(lines) > 1:
            # Multiple lines
            found_zero = False
            new_content = []
            for parts in lines:
                if parts[0] == "0":
                    parts[0] = "2"
                    new_content.append(" ".join(parts) + "\n")
                    found_zero = True
            if not found_zero:
                files_with_label1.append(filename)

        else:
            # Single line
            parts = lines[0]
            if parts[0] == "0":
                parts[0] = "2"
                new_content.append(" ".join(parts) + "\n")
            else:
                files_with_label1.append(filename)

        # Save updated file
        with open(file_path, "w") as f:
            f.writelines(new_content)

print(files_with_label1)
print("Total files with label 1:", len(files_with_label1))
```
