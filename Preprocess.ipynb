```python
import os
from collections import Counter

folder_path = "Potholes\\train\\labels"
label_counts = Counter()

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if parts:  # skip empty lines
                    label = parts[0]
                    label_counts[label] += 1

# Print results
for label, count in sorted(label_counts.items(), key=lambda x: int(x[0])):
    print(f"Label {label}: {count} instances")
```

```python
import os

folder_path = "Potholes\\train\\labels"
files_with_label0 = []

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)
        with open(file_path, "r") as f:
            for line in f:
                parts = line.strip().split()
                if parts and parts[0] == "0":
                    files_with_label0.append(filename)
                    break  # no need to check more lines in this file

print(files_with_label0)
```

```python
import os

txt_folder = "Potholes\\train\\labels"
jpg_folder = "Potholes\\train\\images"

for txt_file in files_with_label0:
    txt_path = os.path.join(txt_folder, txt_file)
    jpg_file = os.path.splitext(txt_file)[0] + ".jpg"
    jpg_path = os.path.join(jpg_folder, jpg_file)
    
    # delete txt
    if os.path.exists(txt_path):
        os.remove(txt_path)
        print(f"Deleted TXT: {txt_file}")
    
    # delete jpg
    if os.path.exists(jpg_path):
        os.remove(jpg_path)
        print(f"Deleted JPG: {jpg_file}")
```
